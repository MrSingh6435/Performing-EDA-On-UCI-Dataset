```python
# First, ensure the necessary libraries are installed
!pip install ucimlrepo pandas seaborn matplotlib

from ucimlrepo import fetch_ucirepo
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Fetch the dataset
heart_disease = fetch_ucirepo(id=45)

# Extract data and targets
X = heart_disease.data.features
y = heart_disease.data.targets

# Inspect column names
print("Feature columns:\n", X.columns)
print("Target column:\n", y.columns)

# Combine the features and target into a single dataframe for easier analysis
data = pd.concat([X, y], axis=1)
print("Combined DataFrame columns:\n", data.columns)

# Step 2: Data Cleaning
# Check for missing values
print("\nMissing values in each column:\n", data.isnull().sum())

# Ensure correct data types
print("\nData types:\n", data.dtypes)

# Step 3: Descriptive Statistics
print("\nDescriptive Statistics:\n", data.describe())

# Step 4: Data Visualization
# Histogram for each feature
data.hist(figsize=(15, 10))
plt.suptitle('Feature Distributions', y=1.02)
plt.tight_layout()
plt.show()

# Pairplot to visualize relationships between features
sns.pairplot(data, hue=y.columns[0])
plt.suptitle('Pairplot of Features', y=1.02)
plt.show()

# Boxplots to check for outliers
plt.figure(figsize=(20, 15))
for i, column in enumerate(data.columns[:-1], 1):
    plt.subplot(4, 4, i)
    sns.boxplot(x=y.columns[0], y=column, data=data)
    plt.title(f'Boxplot of {column}')
plt.tight_layout()
plt.show()

# Correlation heatmap
plt.figure(figsize=(12, 10))
sns.heatmap(data.corr(), annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Heatmap')
plt.show()
```
